```python
df.drop(df.columns[-3:],axis=1,inplace=True)
df.head()
```

```python
locations=df.loc[2][2:]
locations=locations.values
locations
```

```python
units=df['Unnamed: 1'][1:]
units=units.values
units
```

```python
prods=df['SECRETARIA DA AGRICULTURA E DO ABASTECIMENTO – SEAB'][1:]
prods=prods.values
prods
```

```python
empty=['AUS','\\\\','SINF']
df.replace(regex=empty,value=np.NaN,inplace=True)
df.head()
```

```python
df.replace(np.NaN,value='no value',inplace=True)
df.head()
```

```python
prices=df.iloc[1:].copy()
prices=prices[prices.columns[2:]]
prices
```

```python
Location=[]
Productos=[]
Price=[]
Quota=[]
```

```python
len(prods),len(prices)
```

```python
for i in range(prices.shape[1]):
    for j in range(prices.shape[0]):
        Location.append(locations[i])
        Productos.append(prods[j])
        Price.append(prices[prices.columns[i]].values[j])
        Quota.append(units[j])
```

```python
Years=[pd.to_datetime(price_date) for _ in Price]
d={
    'Date':Years,
    'Location':Location,
    'PRODUCTOS':Productos,
    'Price':Price,
    'Quote Type':Quota
}
```

```python
len(Years),len(Location),len(Productos),len(Price),len(Quota),
```

```python
impressao=pd.DataFrame(d)
impressao.head(20)
```
